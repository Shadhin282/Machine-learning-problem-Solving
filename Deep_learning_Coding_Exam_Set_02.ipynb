False
otherwise.
Use this function to print all prime numbers between 2 and 100.

```python
def is_prime(n):
  if n < 2:
    return False
  for i in range(2,int(n**.5)+1):

     if n % i==0 :

      return False

  return True

prime_numbers =  [num for num in range(2, 101) if is_prime(num)]
print(prime_numbers)
```

##Question 7: Capstone Coding Challenge (Bonus Question - Optional) (10 Marks)

### Data Processing and Visualization
Write a Python script that:
Reads a text file
'article.txt'
.
(1 Mark)
Cleans the text by removing punctuation and converting it to lowercase.
(2 Marks)
Counts the frequency of each unique word.
(3 Marks)
Creates a Pandas DataFrame with words and their corresponding frequencies.
(2 Marks)

```python
file=open("/content/drive/MyDrive/Data Set/Deep_learning Data preprocessing data set/Set-02/article.txt","r")
```

```python
file_1 = file.read()
file_1
```

```python
import string
string.punctuation
```

```python
from collections import Counter
remove_punc = [char for char in file_1 if char not in string.punctuation] # Remove punctuation
clean_words = ''.join(remove_punc) # char joining
split_words = clean_words.split()

print(split_words)


```

```python
len(split_words)
```

```python
reviewText = [word for word in split_words if word.lower()]
reviewText
```

```python
word_counts = Counter(reviewText)
word_counts
```

```python
df_text_p = pd.DataFrame(word_counts.items(), columns=['Word', 'Frequency'])

df_text_p
```

```python
df_text_p = df_text_p.sort_values(by='Frequency', ascending=False).reset_index(drop=True)
df_text_p
```

```python
top_15 = df_text_p.head(15)
top_15
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Frequency', y='Word', data=top_15, palette='viridis')
plt.title('Top 15 most Frequent word ')
plt.xlabel('Frequency')
plt.ylabel('Words')
plt.show()
```
